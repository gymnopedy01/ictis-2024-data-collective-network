함수의 매개변수로 함수 전달하기
함수의 매개변수로 전달되는 함수를 콜백함수(callback function)라고 함


```python
#매개변수로 함수에 전달되는 함수 : 콜백
def call_3_times(func):
    for i in range(3):
        func()

def print_hello():
    print('Hello World!')

call_3_times(print_hello)


```

함수를 매개변수로 전달받는 함수의 예
- map() 함수 - map(함수, 리스트)
- filter() 함수 - filter(함수, 리스트)

```python
def power(item):
    return item*item

lst = [1,2,3,4,5]

output = map(power, lst)

print(output)
#print(list(output))

for element in output:
    print(element)
```

```python
def under_3(item):
    return item < 3

lst = [1,2,3,4,5]
output = filter(under_3, lst)
print(output)
#print(list(output))
print(type(output))


for o in output:
    print(o)
```

```python
power = lambda x : x*x

x= range(10)
y= map(power,x)

print(list(y))
```

```python
x= range(10)
y= map(lambda x : x*x,x)

print(list(y))
```

```python
lst = [1,2,3,4,5]
output = filter(lambda x: x<3, lst)

print(list(output))
```

```python
def test():
    print('test() 함수의 첫 줄입니다.')
    try:
        print('try 구문이 실행되었습니다.')
        return
        print('try 구문의 return 키워드 뒤입니다')
    except:
        print('except 구문이 실행되었습니다')
    else:
        print('else 구문이 실행되었습니다.')
    finally:
        print('finally 구문이 실행되었습니다.')
    print('text() 함수의 마지막 줄입니다')

test()

```
